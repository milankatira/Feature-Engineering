<h3 style="color:red">
Mean, Median, Mode imputation

<h3>

```python
def impute_nan(df_target,variable,median):
    df_target[variable+"_median"]=df_target[variable].fillna(median)
```

```python
Median=df_target.LotFrontage.median()
Median
```

```python
impute_nan(df_target,'LotFrontage',Median)
df_target.head()
```

```python
print(df_target['LotFrontage'].std())
print(df_target['LotFrontage_median'].std())
```

```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
fig = plt.figure()

ax = fig.add_subplot(111)

df_target['LotFrontage'].plot(kind='kde', ax=ax)

df_target.LotFrontage_median.plot(kind='kde', ax=ax, color='red')

lines, labels = ax.get_legend_handles_labels()

ax.legend(lines, labels, loc='best')
```

<h3 style="color:red">
Random Sample imputation

<h3>

```python
def impute_nan(df_target,variable,median):
    df_target[variable+"_median"]=df_target[variable].fillna(median)
    df_target[variable+"_random"]=df_target[variable]
```

```python
median=df_target.LotFrontage.median()
```

```python
median
```

```python
impute_nan(df_target,"LotFrontage", median)
```

```python
df_target.head()
```

```python
print(df_target['LotFrontage'].std())
print(df_target['LotFrontage_random'].std())
```

```python
%matplotlib inline
fig = plt.figure()

ax = fig.add_subplot(111)

df_target['LotFrontage'].plot(kind='kde', ax=ax)

df_target.LotFrontage_median.plot(kind='kde', ax=ax, color='red')

df_target.LotFrontage_random.plot(kind='kde' , ax=ax, color='blue')

lines, labels = ax.get_legend_handles_labels()

ax.legend(lines, labels, loc='best')
```

<h3 style="color:red">
End of Distribution imputation

<h3>

```python
df_target=pd.read_csv('titanic.csv')
```

```python
df_target.head()
```

```python
df_target.isnull().sum()
```

```python
df_target.Age.hist(bins=30)
```

```python
sns.boxplot('Age',data=df_target)
```

```python
extreme=df_target.Age.mean()+3*df_target.Age.std()
```

```python
def impute_nan(df, variable, median,extreme):
    df_target[variable+"_end_distribution"]=df_target[variable].fillna(extreme)
    df[variable].fillna(median,inplace=True)
```

```python
impute_nan(df_target,'Age',df_target.Age.median(),extreme)
```

```python
df_target['Age'].hist(bins=30)
```

```python
df_target['Age_end_distribution'].hist(bins=30)
```

```python
sns.boxplot('Age_end_distribution',data=df_target)
```
