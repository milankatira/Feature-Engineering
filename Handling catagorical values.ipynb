```python
def impute_nan(df,variable):
    most_frequant_catagory=df[variable].mode()[0]
    df[variable].fillna(most_frequant_catagory,inplace=True)
```

<h3 style="color:red" > 
Target guided ordinal encoding
<h3>

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
#here target is survived 

df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
ordinal_label=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_label
```

```python
enumerate(ordinal_label,0)
```

```python
ordinal_label2={k:i for i,k in enumerate(ordinal_label,0)}

ordinal_label2
```

```python
df['Cabin_ordinal_label']=df['Cabin'].map(ordinal_label2)

df.head()
```

<h3 style="color:red" > 
Mean encoding
<h3>

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)

df.head()
```

<h3 style="color:red" > 
Probability ratio encoding
<h3>

```python
df=pd.read_csv('titanic.csv',usecols=['Cabin','Survived'])
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
```

```python
prob_df.head()
```

```python
probability_encoding=prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoding']=df['Cabin'].map(probability_encoding)
```
